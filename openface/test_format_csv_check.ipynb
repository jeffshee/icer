```python
csv_dir = "/home/icer/Project/openface_dir/output_simple_vid_data/"
csv_hd_path = csv_dir + "successed_new_head_down.csv"
csv_hu_path = csv_dir + "successed_new_head_up.csv"
csv_hr_path = csv_dir + "successed_new_head_right.csv"
csv_hl_path = csv_dir + "successed_new_head_left.csv"
csv_htr_path = csv_dir + "successed_new_head_tilt_right.csv"
csv_htl_path = csv_dir + "successed_new_head_tilt_left.csv"

df_hd = pd.read_csv(csv_hd_path, engine='python')
df_hu = pd.read_csv(csv_hu_path, engine='python')
df_hr = pd.read_csv(csv_hr_path, engine='python')
df_hl = pd.read_csv(csv_hl_path, engine='python')
df_htr = pd.read_csv(csv_htr_path, engine='python')
df_htl = pd.read_csv(csv_htl_path, engine='python')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['lines.linewidth'] = 0.8
plt.rcParams['lines.markersize'] = 2.0
```

```python
plt.figure()

ax = df_hd.plot(x='timestamp', y='pose_Rx', label='head-down', marker='.')
df_hu.plot(x='timestamp', y='pose_Rx', label='head-up', marker='.', ax=ax, color='C2')
df_hr.plot(x='timestamp', y='pose_Rx', label='head-right', marker='.', ax=ax, color='C3')
df_hl.plot(x='timestamp', y='pose_Rx', label='head-left', marker='.', ax=ax, color='C4')
df_htr.plot(x='timestamp', y='pose_Rx', label='head-tilt-right', marker='.', ax=ax, color='C5')
df_htl.plot(x='timestamp', y='pose_Rx', label='head-tilt-left', marker='.', ax=ax, color='C6', subplots=True)

# plt.savefig('pose_Rx_trial.png', facecolor='white', dpi=300)
# plt.close('all')
```
